#  Netflix EDA Project 

This notebook explores the **Netflix dataset** using Exploratory Data Analysis (EDA).
We will go step by step, explaining each action in simple language, then writing the code, and finally observing the results.


##  Step 1: Import Libraries
  First, we need to import the libraries (tools) we will use for our analysis.
- **pandas**: for handling tables and data
- **numpy**: for numerical operations
- **matplotlib** and **seaborn**: for visualizations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
sns.set(style="whitegrid")
```

## Step 2: Load the Dataset
  Let's load the Netflix dataset from the CSV file and take a quick look at the first few rows.

```python
df = pd.read_csv(r"C:\Users\Gayatri Bhatt\Downloads\netflix_titles.csv")
df.head()
```

## Step 3: Understand Dataset Size and Structure
 Check how many rows and columns are in the dataset, and what types of data we have.

```python
print("Shape (rows, columns):", df.shape)
df.info()
```

##  Step 4: Check Missing Values
 Real datasets often have missing values. Let's see which columns have them.

```python
df.isnull().sum()
```

##  Step 5: Handle Missing Values
To keep things simple, we'll replace missing values with `'Unknown'`.

```python
df['country'] = df['country'].fillna("Unknown")
df['rating'] = df['rating'].fillna("Unknown")
df['date_added'] = df['date_added'].fillna("Unknown")
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
```

##  Step 6: Remove Duplicates
Sometimes the same row is repeated. Let's remove duplicates.

```python
print("Duplicates before:", df.duplicated().sum())
df = df.drop_duplicates()
print("Duplicates after:", df.duplicated().sum())
```

##  Step 7: Descriptive Statistics
Let's generate some descriptive statistics: counts, most frequent values, etc.

```python
df.describe(include='all').T
```

##  Step 8: Movies vs TV Shows
Let's see how much of Netflix is Movies vs TV Shows.

```python
sns.countplot(x="type", data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

##  Step 9: Releases Over the Years
Which years had the most new content released?

```python
df['release_year'].value_counts().sort_index().plot(kind="bar", figsize=(12,5))
plt.title("Number of Releases per Year")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.show()
```

##  Step 10: Top Countries
 Which countries produce the most Netflix content

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind="bar", figsize=(10,5))
plt.title("Top 10 Countries on Netflix")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

## Step 11: Ratings Distribution
 What ratings are most common (e.g., PG, R, TV-MA)?

```python
sns.countplot(y="rating", data=df, order=df['rating'].value_counts().index)
plt.title("Distribution of Ratings")

```

##  Step 12: Movie Durations
How long are Netflix movies usually?

```python
movies = df[df['type']=="Movie"]
movies['duration'] = movies['duration'].str.replace(" min","").astype(float)

movies['duration'].hist(bins=30, figsize=(10,4))
plt.title("Distribution of Movie Durations (minutes)")
plt.xlabel("Minutes")
plt.show()
```

##  Step 13: TV Show Seasons
How many seasons do Netflix TV shows usually have?

```python
tv_shows = df[df['type']=="TV Show"]
tv_shows['duration'].value_counts().sort_index().plot(kind="bar", figsize=(8,4))
plt.title("Number of Seasons (TV Shows)")
plt.xlabel("Seasons")
plt.ylabel("Count")
plt.show()
```

## Step 14: Top Genres
What are the most common genres on Netflix?

```python
df['genre_list'] = df['listed_in'].str.split(", ")
all_genres = df.explode('genre_list')

all_genres['genre_list'].value_counts().head(10).plot(kind="bar", figsize=(10,5))
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
```

## Step 15: Correlation Heatmap
Let's see if numeric features have any relationships.

```python
df['title_length'] = df['title'].str.len()
df['desc_length'] = df['description'].str.len()

num_cols = ['release_year','title_length','desc_length']
plt.figure(figsize=(6,4))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
```

##  Step 16: Variety of Content
How diverse is Netflix’s content?

```python
print("Unique genres:", all_genres['genre_list'].nunique())
print("Unique countries:", df['country'].nunique())
print("Unique ratings:", df['rating'].nunique())
```

##  Step 17: Evolution Over Time
How has the average movie duration changed over decades?

```python
movies['decade'] = (movies['release_year']//10)*10
avg_duration = movies.groupby('decade')['duration'].mean()

avg_duration.plot(kind="line", marker="o", figsize=(8,4))
plt.title("Average Movie Duration by Decade")
plt.xlabel("Decade")
plt.ylabel("Average Duration (minutes)")
plt.show()
```
